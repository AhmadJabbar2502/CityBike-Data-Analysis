```python
import pandas as pd
```

```python
df = pd.read_csv('Data/January/202401-divvy-tripdata.csv')
df.head()
```

```python
df.info()
df.isna().sum()
# df.isnull().sum()
```

```python
df.dropna(subset=['start_station_name', 'end_station_name'], inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
```

```python
df['ride_length_min'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
print(df['ride_length_min'])
```

```python
df['day_of_week'] = df['started_at'].dt.day_name()
df['hour'] = df['started_at'].dt.hour
df['date'] = df['started_at'].dt.date
```
